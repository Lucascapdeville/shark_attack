# Data cleanning dos dados referentes a casos de ataques de tubarão.

Primeira pergunta - Spray anti tubarão aonde vender?

```python
import pandas as pd
import numpy as np
import re
```

```python
shark =  pd.read_csv('data/attacks.csv', sep = ',', encoding = 'windows-1252' )
shark.head(10)
```

```python
shark.columns
```

## Primeiras observações de arquivos:Nome das colunas, limpar valores nones , checar duplicadas.
colunas Unnamed? nan information? pdf?? 

### Arrumando as colunas e removendo as colunas vazias

```python
shark.columns
shark.columns = [col.lower().replace(' ','_').replace('/','_') for col in shark.columns]
shark.drop(['unnamed:_22','unnamed:_23'], axis = 1,inplace = True)
```

```python
shark.head()
```

### Removendo as duplicadas e valores nulos

```python
shark = shark[~shark['case_number'].isna()]
shark = shark[shark['case_number'] != '0']
shark.head(10)
```

# Escolhendo as categorias para analise
colunas: case_number;country;area

```python
shark.columns
```

```python
shark_country = shark.loc[:,['case_number','year','country', 'area','location','sex_']]
shark_country = shark_country[~shark_country['country'].isna()]
print(shark_country['country'].unique())
print(shark_country['country'].value_counts())
```

## USA e Australia são as melhores areas para iniciar a venda do spray de tubarão?
### Aonde desses paises ??

```python
print(shark_country.count())
```

USA representa 35% dos ataques de tubarões
Australia 21% dos ataques de tubaroes

```python
shark_area = shark_country[(shark_country['country'] == 'USA')|(shark_country['country'] == 'AUSTRALIA')]
print(shark_area['area'].value_counts().head(7))
print(shark_area.count())
```

USA - Florida(29%), Hawaii(8%) e California(8%)

Australia -  New south wales(13%), Queensland(8%) e Western Australia(5%)
